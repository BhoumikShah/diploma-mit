```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'A': [23, 45, 56, 78, 34],
    'B': [12, 54, 23, 67, 89],
    'C': [45, 67, 23, 56, 78]
}

df = pd.DataFrame(data)

# 1. Mean, Median, and Mode Calculation
mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]

print("Mean:")
print(mean)
print("\nMedian:")
print(median)
print("\nMode:")
print(mode)

# 2. Standard Deviation and Variance Calculation
std_deviation = df.std()
variance = df.var()

print("\nStandard Deviation:")
print(std_deviation)
print("\nVariance:")
print(variance)

# 3. Correlation Analysis
correlation_matrix = df.corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)

# 4. Descriptive Statistics
descriptive_stats = df.describe()

print("\nDescriptive Statistics:")
print(descriptive_stats)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
data = {
    'A': [23, 45, 56, 78, 34],
    'B': [12, 54, 23, 67, 89],
    'C': [45, 67, 23, 56, 78]
}

df = pd.DataFrame(data)

# Histogram
df.hist()
plt.suptitle('Histogram of Variables', x=0.5, y=1.02, ha='center', fontsize='x-large')
plt.tight_layout()
plt.show()

# Box plot
df.boxplot()
plt.title('Box Plot of Variables')
plt.show()

# Scatter plot matrix
sns.pairplot(df)
plt.suptitle('Scatter Plot Matrix', x=0.5, y=1.02, ha='center', fontsize='x-large')
plt.show()

# Correlation Heatmap
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
